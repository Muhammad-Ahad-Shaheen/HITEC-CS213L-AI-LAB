```python
class Person:
    name = input("Enter Person Name")
    age = int(input("Enter Age"))
class Job:
    designation = input("Enter Person's Post")
    salary = int(input("Enter Salary"))
class Employee(Person,Job):
    def EmployeeInfo(self):
        print(f"Employee Name is: {self.name}")
        print(f"Employee Age is: {self.age}")
        print(f"Employee Designation is: {self.designation}")
        print(f"Employee Salary is: {self.salary}")
emp = Employee()
emp.name
emp.age
emp.designation
emp.salary
emp.EmployeeInfo()
```

```python
class Item:
    basket = []

    def viewFullDescription(self, name, description, price):
        print(f"Item Name: {name}")
        print(f"Item Description: {description}")
        print(f"Item Price: {price}")

    def addToShoppingBasket(self):
        Item.basket.append(self)
        print(f"{self.name} added to basket.")

    def removeFromShoppingBasket(self):
        if self in Item.basket:
            Item.basket.remove(self)
            print(f"{self.name} removed from basket.")
        else:
            print(f"{self.name} not found in basket.")

    @staticmethod
    def viewBasket():
        if not Item.basket:
            print("Your basket is empty.")
        else:
            print("\nðŸ›’ Shopping Basket:")
            total = 0
            for i, item in enumerate(Item.basket, start=1):
                print(f"{i}. {item.name} - ${item.price}")
                total += item.price
            print(f"Total Price: ${total}")

    @staticmethod
    def clearBasket():
        Item.basket.clear()
        print("Shopping basket cleared.")


class MP3(Item):
    def play(self, name, artist, duration):
        print(f"Playing '{name}' by {artist} of Duration {duration}")

    def download(self, name, artist):
        print(f"Downloading '{name}' by {artist}...")


class DVD(Item):
    def viewTrailer(self, name, actor, certificate):
        print(f"Viewing trailer for '{name}' starring {actor} having certificate {certificate}")


class Book(Item):
    def previewContent(self, name, author, numberofpages, genre):
        print(f"Previewing first few pages of '{name}' by {author} having page {numberofpages} and genre {genre}")


while True:
    print("\n--- MENU ---")
    print("1. Add/View Item")
    print("2. Add/View MP3")
    print("3. Add/View DVD")
    print("4. Add/View Book")
    print("5. View Shopping Basket")
    print("6. Clear Shopping Basket")
    print("7. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        item1 = Item()
        item1.name = input("Enter Item Name: ")
        item1.description = input("Enter Item Description: ")
        item1.price = int(input("Enter Item Price: "))
        item1.viewFullDescription(item1.name, item1.description, item1.price)
        action = input("Add to basket? (yes/no): ").lower()
        if action == "yes":
            item1.addToShoppingBasket()

    elif choice == "2":
        mp3_1 = MP3()
        mp3_1.name = input("Enter MP3 Name: ")
        mp3_1.price = int(input("Enter MP3 Price: "))
        artist = input("Enter Artist Name: ")
        duration = input("Enter Duration: ")
        mp3_1.play(mp3_1.name, artist, duration)
        mp3_1.download(mp3_1.name, artist)
        action = input("Add to basket? (yes/no): ").lower()
        if action == "yes":
            mp3_1.addToShoppingBasket()

    elif choice == "3":
        dvd_1 = DVD()
        dvd_1.name = input("Enter DVD Name: ")
        dvd_1.price = int(input("Enter DVD Price: "))
        actor = input("Enter Actor Name: ")
        certificate = input("Enter Certificate: ")
        dvd_1.viewTrailer(dvd_1.name, actor, certificate)
        action = input("Add to basket? (yes/no): ").lower()
        if action == "yes":
            dvd_1.addToShoppingBasket()

    elif choice == "4":
        book_1 = Book()
        book_1.name = input("Enter Book Name: ")
        book_1.price = int(input("Enter Book Price: "))
        author = input("Enter Author Name: ")
        pages = int(input("Enter Number of Pages: "))
        genre = input("Enter Genre: ")
        book_1.previewContent(book_1.name, author, pages, genre)
        action = input("Add to basket? (yes/no): ").lower()
        if action == "yes":
            book_1.addToShoppingBasket()

    elif choice == "5":
        Item.viewBasket()

    elif choice == "6":
        Item.clearBasket()

    elif choice == "7":
        print("Exiting...")
        break

    else:
        print("Invalid choice! Please try again.")
```

```python
class power:
    def pow(self,x,n):
        if n==0:
            return 1
        elif n>0:
            return x**n
        else:
            result = x**abs(n)
            return 1/result
p=power()
num = int(input("Enter A Number"))
power = int(input("Enter to the Power"))
print(f"{num} raised to the Power {power} is: {p.pow(num,power)}")
```

```python
class Circle:
    def circleArea(self,radius):
        area = 3.1415*radius*radius
        return area
    def circleCircumference(self,radius):
        circum = 2*3.1415*radius
        return circum
c = Circle()
radius = float(input("Enter Radius of Circle"))
print(f"Area of Circle is: {c.circleArea(radius)}")
print(f"Circumference of Cricle is: {c.circleCircumference(radius)}")

```
